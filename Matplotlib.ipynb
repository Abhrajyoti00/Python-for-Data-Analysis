```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
plt.plot(x,y)
plt.xlabel('X LABEL')
plt.ylabel('Y LABEL')
plt.title('Title')
```

```python
plt.subplot(1,2,1)  #Row,column
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#OO
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8]) #left,bottom,width,height... 0 to 1

axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')

axes.set_title('Set Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)
axes2.set_title('Y vs X')
axes1.set_title('X vs Y ')
```

```python
fig,axes = plt.subplots(nrows =1, ncols =2)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows =1, ncols =2)

axes[0].plot(x,y)
axes[0].set_title('X vs Y')
axes[1].plot(y,x)
axes[1].set_title('Y vs X')

plt.tight_layout()
```

### Figure Size and DPI

```python
fig = plt.figure(figsize=(8,2)) #Figsize = Tuple of Width and Height in inches
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,ax = plt.subplots(nrows =2, ncols = 1, figsize=(8,2)) #Figsize = Tuple of Width and Height in inches

ax[0].plot(x,y)
ax[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('my_picture.png',dpi=200)
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2)
axes.plot(x,x**3)
axes.set_title("Title")
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2, label = 'X Sqd')
axes.plot(x,x**3, label = 'X Cubed')
axes.set_title("Title")
axes.legend(loc = 0)
```

# Plot Appearance

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='red',linewidth=3,alpha=0.5)  #Alpha = transparency
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='red',lw=1,linestyle='--',marker='o')  
```
